### Contextualización histórica  
Los contratos forward tienen raíces milenarias. Ya en el siglo XVIII a.C., el Código de Hammurabi reglamentaba acuerdos para entrega futura de bienes, estableciendo precios y fechas de liquidación (el “first forward contract”) en la antigua Mesopotamia  ([History of derivatives, Part 2: Forward contracts - Olymp Trade Blog](https://blog.olymptrade.com/education/history-of-derivatives-part-2?)).
Durante la Antigua Grecia y Roma se utilizaron instrumentos semejantes para gestionar riesgos agrícolas y comerciales. 
En la Edad Media europea, ferias mercantiles en Italia y Flandes formalizaron compromisos forward sobre textiles y granos. 
Finalmente, en 1864 la Chicago Board of Trade estandarizó estos contratos, luego evolucionando, a través los futuros con clearing y mitigación parcial o total del riesgo de contraparte  ([Chicago Board of Trade](https://en.wikipedia.org/wiki/Chicago_Board_of_Trade)).

### Glosario
- $S_0$: precio spot del subyacente al inicio.  
- $S_T$: precio spot en la fecha de vencimiento $T$.  
- $F_0(T)$: precio pactado del forward al inicio para entrega en $T$.  
- $r$: tasa libre de riesgo (continua o discreta).  
- $T$: tiempo hasta vencimiento (en años).  
- $V_t(T)$: valor de mercado (MTM) del forward en el instante $t$.

## 1. Características y riesgos  
Los forwards son contratos **bilaterales**, **personalizables** (cantidad, fecha, activo) y sin pago inicial . Al ser OTC, conllevan:
- **Riesgo de contraparte**: posibilidad de default de la contraparte.  
- **Riesgo de Precios/Tasas**: variaciones adversas del precio del activo subyacente $S_t$.  
- **Riesgo de liquidez**: dificultad para cerrar o modificar la posición con un costo adicional.
- **Riesgo de base**: divergencia entre la variación del derivado y la del subyacente.  
- **Riesgo sistémico**: El riesgo de que el uso de derivados en el mercado y los riesgos excesivos produzcan eventos de estrés, contagio financiero y/o pánico en los mercados.


### Funcionamiento de los forwards  
Un forward es un **acuerdo OTC** entre dos partes para comprar/vender un activo a *Precio forward* $F_0(T)$ en $T$, sin intercambio de efectivo inicial y con payoff lineal:
$$
V_T(T) = S_T - F_0(T).
$$
La ausencia de pago inicial hace que $V_0(T)=0$ y, al vencimiento, el comprador gana si $S_T>F_0(T)$ y pierde en caso contrario .


## 2. Gobierno y normativa  

### 2.1 Gobernanza y normativa internacional  

En el entorno de derivados, tanto las normas contables como los requerimientos de capital bancario definen **cómo** se registran, miden y respaldan estos contratos frente a riesgos financieros y de contraparte.

- **IFRS 9**  
  - **Clasificación y medición**: todos los derivados se reconocen a **fair value**, con variantes de presentación (FVPL vs. FVOCI) según su modelo de negocio y objetivo de cobertura.  
  - **Hedge accounting**: permite mitigar la volatilidad en P&L mediante tres tipos de coberturas (fair value, cash flow, net investment), sujeto a requisitos de documentación y efectividad.  
  - **Pérdidas esperadas**: incorpora el enfoque de **Expected Credit Loss** para evaluar exposición por contrapartes y, de manera indirecta, el riesgo de crédito derivado.

- **Basilea III**  
La serie de Acuerdos de Basilea surge como respuesta a las crisis financieras que revelaron debilidades en la regulación bancaria y la gestión de riesgos de contraparte—especialmente en derivados—y busca reforzar la resistencia del sistema financiero global.  

**Orígenes históricos y motivación**  
Tras la implementación de Bailea I, se estableció un requerimiento mínimo de capital del 8 % de los activos ponderados por riesgo, y las mejoras de **Basilea II** (2004), quedó en evidencia durante la crisis subprime de 2008 que los bancos carecían de capital suficiente y de liquidez para absorber pérdidas inesperadas, y que las métricas existentes no capturaban plenamente el riesgo de contraparte en derivados ni el riesgo sistémico. En respuesta, el Comité de Supervisión Bancaria de Basilea (BCBS) promulgó **Basilea III** en 2010, con fases de implementación hasta 2019 (y “Endgame” extendido a 2025), para remediar tres deficiencias clave:  
1. **Requerimientos de capital**: calidad y cantidad de capital propias, en base a los activos ponderados por riesgo.
2. **Liquidez inadecuada**: falta de activos líquidos de alta calidad para sobrevivir a tensiones de corto y mediano plazo.  
3. **Exposición CCR (Counterparty Credit Risk)**: subestimación de pérdidas potenciales por quiebra de contrapartes, especialmente en derivados OTC
4. **Ratio de apalancamiento**: introduce un límite no ponderado por riesgo (3 %) para desalentar endeudamiento excesivo.
5. **Ratios de liquidez**:

   - LCR (Liquidity Coverage Ratio): exige cobertura de salidas netas de efectivo a 30 días con High-Quality Liquid Assets.

   - NSFR (Net Stable Funding Ratio): promueve financiación estable a un horizonte de 1 año.

## 3. Beneficios y aplicaciones  
Los forwards permiten diversas aplicaciones prácticas:

1. **Cobertura corporativa**: un importador de insumos puede fijar el tipo de cambio para pagar en moneda extranjera (Libra esterlina GBP) sin **riesgo de tipo de cambio**.  
   *Ejemplo*: Montau AG vende una máquina en GBP y asegura EUR via un forward GBP/EUR .  

2. **Especulación**: traders de índices monitorizan forwards FX o forwards sobre commodities para anticipar movimientos y obtener ganancias sin inmovilizar capital.  
   *Ejemplo*: observar el forward de S&P 500 antes de la apertura, explotando diferencias temporales de precio .  
3. **Arbitraje convencional**: aprovechar diferencias de tasas entre dos activos (accion o bono del mismo emisor con igual madurez) a precios distintos, por ejemplo vender en corto a un precio más alto un activo A  (99,15) y comprar el mismo activo en mercado spot a precio más bajo B (99). Al fin del periodo de inversión los activos debieran costar lo mismo (A=B 100=100). En este punto se puede vender A a 100 y comprar B a 100 para cerrar la venta corta obteniendo 0,15 libre de riesgo.
    *Ejemplo*: Basis Spread.

4. **Arbitraje covered interest**: aprovechar diferencias de tasas entre dos monedas: comprar spot, financiar a $r_{\text{local}}$, vender forward y capturar el diferencial de tasas entre 2 economías 

    $e^{(r_{\text{local}} - r_{\text{extranjero}})T}$.  
    *Ejemplo*: Carry trade ("Carry" o "Holding" deriva desde mantener el activo para concretar el arbitraje.)

5. **Gestión de liquidez**: generar exposición sin comprar el activo subyacente en el mercado spot, reduciendo requerimientos de caja inicial y costos de transacción. Mercados más liquidos y ajuste de plazos -> Mayor liquidez para el negocio y mejorar EBITDA.



## 4. Arbitraje y mercados eficientes  
Bajo el principio de no-arbitraje se cumple:
$$
F_0(T) = S_0 e^{rT}
$$
(en discreto, $F_0(T)=S_0(1+r)^T$) .  


Mediante **replicación**, se muestra que la combinación de una posición spot y financiamiento a $r$ reproduce el payoff de un forward, eliminando oportunidades de arbitraje .



## 5. Réplica

1. **Ir long en el forward** (comprar USD a futuro a 950 CLP/USD)  
2. **Vender USD hoy en spot (a 970 CLP/USD)**  
   - Esto te entrega CLP hoy.
   - Lo inviertes a la tasa chilena (8%).

3. **Al vencimiento**, compras los USD al precio pactado (950) usando los CLP que invertiste a 8%.

Esto te da una ganancia libre de riesgo por la diferencia entre el valor que debieras pagar (según tasas) y lo que efectivamente pagarás (el precio de ejercicio bajo).

$$
F_0(T) = S_0*(1 + r)^T
$$
This relationship is shown under continuous compounding:

$$
F_0(T) = S_0*e^{rT}
$$



So... in the case of currencys

$$
F_0(T) = S_0*e^{(r_l - r_e)T}
$$


## 6. Ejemplo aplicado al caso chileno CLP / USD  

*Puedes revisar la información real de precios en la página oficial del Banco Central de Chile*.
[Precios históricos de FWD CLP/USD](https://si3.bcentral.cl/siete/ES/Siete/Cuadro/CAP_DYB/MN_ESTAD_MON55/EM_OMA_VF_28?idSerie=F022.VFWP4.PRE.D028.Z.USD.D).

Podemos resumir esta relación como invertir a tasa spot local y descontar el flujo del pago de intereses en la tasa foránea y ajustarlo por el horizonte de inversión.

Esto nos da un resultado donde el inversor es indiferente.

![payoffs replication.png](<attachment:payoffs replication.png>)

```python
import numpy as np
import matplotlib.pyplot as plt

def forward_theoretical_price(S_t: float, r_domestic: float, r_foreign: float, T: float) -> float:
    """
    Calcula el precio teórico de un forward usando paridad de tasas de interés (Interest Rate Parity).

    Args:
        S_t (float): Precio spot actual.
        r_domestic (float): Tasa de interés de la moneda nacional (CLP).
        r_foreign (float): Tasa de interés de la moneda extranjera (USD).
        T (float): Tiempo al vencimiento en años.

    Returns:
        float: Precio teórico del forward.
    """
    return S_t * ((1 + r_domestic) / (1 + r_foreign))**T

def payoff_forward(S_T: float, K: float, N: float, position: str = "long") -> float:
    """
    Calcula el payoff de un contrato forward.

    Args:
        S_T (float): Precio spot al vencimiento.
        K (float): Precio de ejercicio del forward.
        N (float): Notional del contrato en USD.
        position (str): 'long' o 'short'.

    Returns:
        float: Payoff en CLP.
    """
    payoff = N * (S_T - K)
    return payoff if position == "long" else -payoff


# Parámetros base del contrato forward
spot_price = 970  # S_t
strike_price = 950  # K
notional_usd = 1000  # N
r_clp = 0.08  # tasa en Chile (8%)
r_usd = 0.05  # tasa en USA (5%)
T_days = 360
T_years = T_days / 360  # base 360

# Calcular precio forward teórico
F_teorico = forward_theoretical_price(spot_price, r_clp, r_usd, T_years)

# Simulación de payoff
spot_range = np.linspace(900, 1000, 100)
payoff_long = [payoff_forward(S, strike_price, notional_usd, position='long') for S in spot_range]
payoff_short = [payoff_forward(S, strike_price, notional_usd, position='short') for S in spot_range]
```

```python

# Gráfico
plt.figure(figsize=(10, 6))
plt.plot(spot_range, payoff_long, label='Long Forward', linestyle='--')
plt.plot(spot_range, payoff_short, label='Short Forward', linestyle='-.')
plt.axvline(strike_price, color='gray', linestyle=':', label='Strike Price (K)')
plt.axhline(0, color='black', linewidth=0.5)
plt.title("Payoff (P&L) del Contrato Forward CLP/USD")
plt.xlabel("Precio Spot al vencimiento (S_T)")
plt.ylabel("Payoff (CLP)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print(f"F_teorico: {round(F_teorico,4)} CLP/USD")
```


#### 🧠 ¿Qué significa esta diferencia?

Relación entre el **precio teórico** y el **precio observado en el mercado**:

- **Precio teórico > Precio observado**  
  → El forward **está subvalorado** respecto a su valor justo.  
  → El mercado te está ofreciendo comprar USD a futuro (a 950 CLP/USD) más barato que el valor que se justificaría por diferencias de tasas de interés (997.71 CLP/USD).

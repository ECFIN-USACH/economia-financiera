### Pricing (Precio Forward)

El **precio forward** $F_0(T)$ es el monto acordado hoy para intercambiar un activo en una fecha futura $T$. Este precio se determina bajo el principio de **no arbitraje**: debe ser equivalente a replicar la inversiÃ³n mediante la compra spot financiada a la tasa libre de riesgo.

Para un activo sin flujos intermedios:

$$

$$

Donde:

* $S_0$: Precio spot
* $r$: Tasa libre de riesgo (riesgo neutral)
* $T$: Tiempo hasta el vencimiento



### 2. Valuation (Valor de un Forward)

El **valor de un contrato forward** en el tiempo $t$, antes de su vencimiento $T$, es:

$$
V_t(T) = S_t - F_0(T)(1 + r)^{-(T - t)}
$$

* Si el precio spot ha subido, el comprador (posiciÃ³n larga) gana valor.
* Si el spot ha bajado, el contrato tiene valor negativo para el comprador.

### 3. Forward con Costos o Beneficios (Cost of Carry)

Cuando el activo subyacente **genera flujos (beneficios $I$)** o incurre en **costos ($C$)**, el precio forward se ajusta:

$$

$$

* Los ingresos como **dividendos** se descuentan.
* Los costos como **almacenaje, seguros, custodia** se agregan.

$$

### 3. Forward Rate Agreements (FRA)

Un **Forward Rate Agreement (FRA)** es un contrato derivado **OTC (Over-the-Counter)** que permite a dos contrapartes **fijar hoy una tasa de interÃ©s para un perÃ­odo futuro**, sobre un monto nocional que no se intercambia.

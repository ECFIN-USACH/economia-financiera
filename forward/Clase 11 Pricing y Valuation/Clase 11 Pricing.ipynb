## Clase 11 - Pricing y Valuation de Contratos Forward

---


## 🎯 Objetivos de Aprendizaje

Al finalizar esta clase, el estudiante será capaz de:

1. Distinguir entre el **precio** y el **valor** de un contrato forward.
2. Valorar forwards en distintos momentos del tiempo, con o sin costos/beneficios asociados.
3. Entender y aplicar forwards sobre activos financieros con estructura temporal (spot vs. forward).
4. Comprender el funcionamiento de **Forward Rate Agreements (FRA)** como derivados de tasas de interés.




### 1. Pricing vs. Valuation de un Contrato Forward

* **Precio del forward (F₀(T))**: Es el precio pactado al inicio del contrato para transar el activo en el futuro.
* **Valor del forward (Vₜ(T))**: Es el valor de mercado del contrato en un momento t antes de su vencimiento, dado por la diferencia entre el valor actual del activo y el valor presente del precio forward.

🔹 **¿Por qué es importante?** Aunque el contrato inicia con valor cero, **el valor puede cambiar con el tiempo**, generando ganancias o pérdidas latentes según cómo varíe el precio del activo subyacente o la tasa libre de riesgo.



### Pricing (Precio Forward)

El **precio forward** $F_0(T)$ es el monto acordado hoy para intercambiar un activo en una fecha futura $T$. Este precio se determina bajo el principio de **no arbitraje**: debe ser equivalente a replicar la inversión mediante la compra spot financiada a la tasa libre de riesgo.

Para un activo sin flujos intermedios:

$$
F_0(T) = S_0(1 + r)^T
$$

Donde:

* $S_0$: Precio spot
* $r$: Tasa libre de riesgo (riesgo neutral)
* $T$: Tiempo hasta el vencimiento



### 2. Valuation (Valor de un Forward)

El **valor de un contrato forward** en el tiempo $t$, antes de su vencimiento $T$, es:

$$
V_t(T) = S_t - F_0(T)(1 + r)^{-(T - t)}
$$

* Si el precio spot ha subido, el comprador (posición larga) gana valor.
* Si el spot ha bajado, el contrato tiene valor negativo para el comprador.



### Ejemplo – Valorización con activo sin flujos

**AMY Investments** acuerda comprar acciones de Airbus a EUR 30 en un año. Supón que hoy el spot es EUR 29.70, y $r = 1\%$.
El forward es:

$$
F_0(T) = 29.70 \cdot (1.01)^1 = 30.00
$$

Después de 6 meses, si el precio spot sube a 31.00:

$$
V_{0.5}(T) = 31.00 - 30.00 \cdot (1.01)^{-0.5} = \text{Ganancia para el comprador}
$$



### 3. Forward con Costos o Beneficios (Cost of Carry)

Cuando el activo subyacente **genera flujos (beneficios $I$)** o incurre en **costos ($C$)**, el precio forward se ajusta:

$$
F_0(T) = [S_0 - PV(I) + PV(C)](1 + r)^T
$$

* Los ingresos como **dividendos** se descuentan.
* Los costos como **almacenaje, seguros, custodia** se agregan.

Para el caso de capitalización continua utilizamos euler:

$$
F_0(T) = [S_0]*e^{(r+c-i)*T}
$$

---



###  Ejemplo – Acciones con Dividendos




**Hightest Capital** firma un forward para entregar 1,000 acciones de Unilever a EUR 50.6311 por acción a 6 meses. La acción se espera que pague dividendos de EUR 0.30 en 2 meses. (Para este caso se puede en una versión simplificada estimar la tasa de dividendo a partir de datos históricos)
Tasa libre de riesgo $r = 5\%$.

Calculemos el **Precio** Spot que permite generar un MTM del derivado igual a 0 (Forward Contract Breakeven).

Se calcula el valor presente del dividendo:

$$
PV(I) = 0.30(1.05)^{-2/12} = 0.2976
$$

A 4 meses, el precio spot de equilibrio (donde $V_t(T) = 0$) es:

$$
S_t = F_0(T)(1 + r)^{-(T - t)} + PV(I) = 50.2202 + 0.2976 = 50.5178
$$

### 3. Forward Rate Agreements (FRA)

Un **Forward Rate Agreement (FRA)** es un contrato derivado **OTC (Over-the-Counter)** que permite a dos contrapartes **fijar hoy una tasa de interés para un período futuro**, sobre un monto nocional que no se intercambia.

Esencialmente, se trata de una **apuesta estructurada sobre la evolución futura de las tasas de interés**, y es ampliamente utilizado para cobertura (hedging) o especulación por bancos, gestores de fondos y corporaciones.

#### Estructura del FRA

* **Fijación del contrato (t = 0):**

  * Se acuerda:

    * Fecha de inicio del período futuro $t = A$
    * Duración del período $B - A$
    * Tasa fija acordada: $\text{IFR}_{A,B-A}$ (forward rate implícita)
    * Monto nocional (no intercambiado)
    * Convención de días: usualmente **actual/360** o **actual/365**

* **Liquidación (t = A):**

  * La tasa variable de referencia (**MRR**, Market Reference Rate) se observa en el mercado.
  * Se calcula el diferencial entre la tasa flotante y la tasa fija pactada.
  * El resultado neto se **liquida en efectivo al inicio del período** $[A, B]$, con descuento al valor presente.




#### Interpretación Económica

* El **comprador del FRA (posición larga)** paga tasa fija y recibe tasa flotante. Gana si las tasas suben.
* El **vendedor del FRA (posición corta)** recibe fija y paga flotante. Gana si las tasas bajan.

---
### Ejemplo Detallado – FRA en CNY (Caso Yangzi Bank)

**Contexto:**
Yangzi Bank busca protegerse ante el riesgo de alza en tasas. Para ello, **paga tasa fija** en un FRA que cubre un período de 3 meses que comienza en 3 meses más.

* **Notional:** CNY 100,000,000
* **Tasa fija (IFR):** 2.24299% anual (base 360)
* **Período de cobertura:** 90 días (0.25 años)
* **Tasa flotante observada al inicio (MRR):** 2.15%


#### 1️⃣ Paso 1 – Cálculo del pago neto nominal

$$
\text{Pago Nominal} = (\text{MRR} - \text{Fijo}) \cdot \text{Notional} \cdot \frac{D}{360}
$$

$$
= (0.0215 - 0.0224299) \cdot 100{,}000{,}000 \cdot \frac{90}{360} = -CNY 23{,}247.50
$$

🔻 Como la tasa de mercado es **menor que la tasa fija**, Yangzi Bank pierde (debe pagar).


#### 2️⃣ Paso 2 – Ajuste por valor presente (liquidación anticipada)

El pago se liquida **al inicio del período de cobertura**, por lo que debe descontarse al presente usando la tasa flotante observada $\text{MRR} = 2.15\%$:

$$
\text{PV} = \frac{\text{Pago Nominal}}{1 + \frac{\text{MRR} \cdot D}{360}} = \frac{-23{,}247.50}{1 + \frac{0.0215 \cdot 90}{360}} = \frac{-23{,}247.50}{1.005375} = -CNY 23{,}123.21
$$


#### 💡 Interpretación Económica

* Yangzi Bank paga **CNY 23,123.21** al inicio del trimestre.
* Sin embargo, este monto **queda compensado por un menor costo financiero** si su pasivo real está indexado a MRR (por ejemplo, un préstamo en tasa variable o bono flotante).
* El FRA actúa como **un seguro contra alzas de tasas**: si las tasas hubiesen subido por encima del 2.24299%, Yangzi habría ganado.

## 1. Introducción e Importancia  
Los derivados son **instrumentos esenciales** en el sistema financiero. Se basan en generar un contrato basado en el valor de un activo subyacente. Su uso permite:  
- **Transferir riesgo** entre agentes (tipo de cambio, tasa de interés, commodities).  
- **Mejorar la eficiencia** de precios y liquidez en mercados organizados y OTC.  
- **Optimizar el capital**, al ofrecer exposiciones apalancadas con menor desembolso inicial.  

En Chile, el mercado de derivados ha crecido fuertemente desde la creación de la Bolsa Electrónica de Chile y la CMF supervisa hoy contratos OTC y estandarizados.  

  


### Principales diferencias entre **renta fija**, **acciones** y **derivados**:

| Característica              | Renta Fija                           | Acciones                              | Derivados                                         |
|-----------------------------|--------------------------------------|---------------------------------------|---------------------------------------------------|
| **Naturaleza del instrumento** | Título de deuda                     | Título de propiedad                  | Contrato entre dos partes                         |
| **Flujo de caja**           | Cupones periódicos + devolución de principal al vencimiento | Dividendos variables + posible recompra de acciones | Ninguno hasta ejercicio/liquidación (excepciones opcionales) |
| **Titularidad**             | Acreedor                             | Copropietario (equity)                | No confiere propiedad; expone al subyacente       |
| **Plazo**                   | Definido (fijo o amortizable)        | Indefinido (perpetua mientras existe la empresa) | Definido por contrato                             |
| **Requerimiento inicial**   | Precio de emisión o mercado          | Precio de mercado                     | Margen inicial (futuros/opciones) o pago de prima |
| **Liquidación**             | Pago de cupones y principal          | Transferencia de acciones             | Entrega física o cash settlement según contrato   |

## 2. Características, Tipos y Usos  
### 2.1. Estructura contractual  



- **Forward Commitments (compromisos):** obligaciones bilaterales de comprar/vender en el futuro a precio acordado.  
  - **Forwards OTC**: flexibles, sin cámara de compensación.  
    - **Tipo de entrega:** habitualmente física, aunque puede pactarse liquidación en efectivo. se acuerda lugar y modo de entrega.  
    - **Settlement:** mark-to-market según contabilidad, el P&L se liquida al vencimiento.  
  - **Futures**: Se transan en el mercado secundario, con colateral gestionado por cámaras de compensación.  
    - **Tipo de entrega:** puede ser física (entrega del activo) o cash settlement.  
    - **Settlement:** diario por mark-to-market; liquidación final en efectivo o física según contrato.  
  - **Swaps**: serie de forwards implícitos (swap de tasas (*IRS*), FX swap (*CCS*, swap de crédito (*CDS*)) OTC.  
    - **Tipo de entrega:** siempre cash settlement, intercambiando flujos de caja periódicos. No tienen entrega física.
    - **Settlement:** netting de flujos en fechas predeterminadas.  

- **Contingent Claims (derechos contingentes):** opciones que dan derecho, pero no la obligación, a ejercer el contrato.  
  - **Options** Europeas, americanas, bermuda, etc.  
    - **Tipo de entrega:** física (entrega/subyacente) o cash settlement.  
    - **Settlement & Delivery:** Americanas pueden ejercerse en cualquier momento; europeas sólo al vencimiento.  



### 2.2. Finalidades económicas  
1. **Cobertura (hedge):**  
   - Eliminar riesgo de precio/incertidumbre (p.ej. DV01 negativo para inmunizar cartera de bonos), Risk Management, posicionamiento en la curva.

2. **Apalancamiento:**  
   - Exposición multiplicada: un contrato futuro de tasa puede controlar montos mucho mayores que el margen inicial, gracias a la modalidad de compensación diaria.  

3. **Especulación:**  
   - Buscar ganancias vía delta-neutral trading, volatilidad implícita, Spread convergence/divergence, etc.  

4. **Arbitraje:**  
   - Explotar desalineaciones entre spot y forward, o entre mercados (copper futures Codelco).  
   - Carry trade.

5. **Costos de transacción**
    - Se utilizan derivados para operar en el *mercado spot* (commodities por ejemplo) para vender o comprar el activo subyacente. Generalmente está relacionado a una mayor liquidez de mercado.


### 2.3. Elementos comunes  
- **Activo subyacente:** FX, equity, bonos, commodities, tasas, índices de mercado, clima, default crediticio, etc.  
- **Precio de ejercicio (K)** y **fecha de expiración o vencimiento (T):** precio acordado y momento de finalización del contrato.  
- **Liquidación:**  
  - **Entrega física:** transmisión del activo subyacente al vencimiento.  
  - **Cash settlement:** pago de la diferencia en efectivo según precio de referencia.  
- **Margen y colateral:**  
  - **Initial Margin:** depósito de garantía al abrir posición (futuros).  
  - **Variation Margin:** ajustes diarios para reflejar el mark-to-market.  
  - **Colateral:** efectivo, valores de alta liquidez o líneas de crédito.  
- **Participantes y contrapartes:**  
  - **Long Holder:** se beneficia por un alza en el subyacente (*posición larga*).  
  - **Short Holder:** se beneficia por una baja en el subyacente (*posición corta*).  
  - **Market Makers:** cotizan bid-ask, asumen inventario y proporcionan liquidez continua.  
  - **Bancos:** estructuran y venden derivados OTC, gestionan colateral y ofrecen líneas de crédito a clientes.  
  - **Cámaras de Compensación (*Central counterparty*) o Clearing Houses:** interponen una contraparte central mitigando riesgo de crédito, mediante el proceso de liquidación y settlement, gestionan garantías y llamadas de margen.  
  - **Reguladores e instituciones internacionales(CMF, Banco Central, Basilea III, IFRS9)**

## 3. Arbitraje y Breve Historia
### 3.1. Principio de no-arbitraje  

> Un **modelo libre de arbitraje** asegura que no existan oportunidades de ganancia sin riesgo. De aquí derivan las ecuaciones de precios de forwards y opciones.

**Forward plain-vanilla:**  
$$
F(0,T) = S_0 \, e^{(r - q)\,T}
$$  
donde $S_0$ es el spot, $r$ tasa libre de riesgo y $q$ yield del activo (dividendos, convenience yield).  





### 3.2. Hitos históricos  
1. **Tales de Mileto (c. 600 a.C.)**: contrato de uso de prensa de aceite de oliva para asegurar precio ante una expectativa de buenas cosechas de aceitunas para el verano en base a pronóstico climático.



2. **Sir Isaac Newton (1687)**: desarrollo del cálculo diferencial, base del trato con flujos infinitesimales (Procesos estocásticos, entre otras aplicaciones). En 1720 Newton perdió aproximadamente 1/3 de su fortuna en los mercados financieros.





3. **Louis Bachelier (1900)**: en *Théorie de la spéculation* modela precios con movimiento browniano → antecesor del modelo Black–Scholes. A través de la radiación de probabilidades genera un modelo para valorizar opciones.

## 4. Pricing, Valuation, Mercado Actual y Regulaciones  
### 4.1. Pricing vs. Valuation  
- **Pricing**: cálculo matemático del precio de negociación en t=0, basándose en no-arbitraje y tasas de financiamiento.  
- **Valuation**: estimación contable (IFRS 9, NIIFs locales) del valor razonable a reportar.  

### 4.2. Modelos de pricing  
- **Forwards/Futuros**: modelo de carry trade de tasas de interés.  
- **Opciones**: Black–Scholes–Merton para europeas sobre acciones; modelos de volatilidad local/stocástica para mercados modernos.  
$$
C = S_0\,N(d_1) - K\,e^{-rT}\,N(d_2)
$$  
con  
$\displaystyle d_{1,2} = \frac{\ln(S_0/K) + (r \pm \tfrac12\sigma^2)T}{\sigma\sqrt{T}}$.

### 4.3. Mercado chileno  
- **OTC**: forwards y swaps bajo supervisión CMF, con contratos bilaterales.  
- **Exchange–Traded**: futuros de cobre, futuros USD/CLP, opciones equity, respaldados por cámara de compensación.  

### 4.4. Regulaciones clave  
- **CMF** exige reporte de posiciones, márgenes y cálculo de capital regulatorio según Basilea III.  
- **Normas contables**: reconocimiento a valor razonable *Fair-Value* bajo IFRS 9.  
- **Gestión de colateral**: Contratos OTC, requerimientos de capital normativo, clearing houses, margin call.





## 5. Riesgos y Estrategias: Análisis de Sensibilidad  
### 5.1. Riesgos principales  
- **Mercado:** variaciones de precio, tasas y volatilidad.  
- **Crédito:** Incumplimiento de obligaciones crediticias. XVA's.  
- **Liquidez:** no poder cerrar posición sin costo extra.
- **Operacional:** errores de ejecución, riesgo reputacional, etc.  
- **Modelos:** Model Risk Management, riesgo de que los modelos se desajusten en cualquier aspecto (data, calculo, desajuste de parámetros, etc.)

### 5.2. Métricas de sensibilidad  
#### Para forwards/futuros  
- **DV01** (Dollar Value of 01): cambio en valor por 1 pb de tasa.  
$$
\text{DV01} \approx -\frac{\Delta P}{\Delta y}
$$  
- **Duration** y **Convexity** para swaps.  

#### Para opciones (“Greeks”)  
- **Delta**: $\partial V/\partial S$  
- **Gamma**: $\partial^2V/\partial S^2$  
- **Vega**: $\partial V/\partial \sigma$  
- **Theta**: $\partial V/\partial T$  
- **Rho**: $\partial V/\partial r$

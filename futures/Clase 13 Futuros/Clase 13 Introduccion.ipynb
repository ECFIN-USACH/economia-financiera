#### 1. Principales caractersticas de los contratos de futuros

 **Estandarizacin y negociacin en bolsa organizada**
 Cada contrato de futuros especifica de antemano el activo subyacente (p. ej., barriles de petrleo Brent, toneladas mtricas de trigo, divisa EUR/USD), el tamao del lote (e.g., 1 lote = 1 000 bushels de maz), la calidad o grado mnimo del activo, las fechas de vencimiento (normalmente uno o varios meses fijos al ao) y el mecanismo de liquidacin (fsica o cashsettlement). Esta estandarizacin, impuesta por la bolsa (p. ej., CME, ICE, Euronext), permite que cualquier participante pueda entrar o salir de una posicin sin necesidad de negociar condiciones adicionales.

 **Intervencin de la cmara de compensacin (clearing house)**
 La cmara de compensacin acta como contraparte central para todas las transacciones: cuando alguien abre posicin compradora, la cmara se convierte en su contraparte vendedora, y viceversa. Esto elimina el riesgo directo de impago entre dos contrapartes cualquiera, ya que la cmara garantiza el cumplimiento incluso si alguna de las partes originales no pudiera cubrir sus obligaciones.

 **Liquidacin diaria (mark to market)**
 A diferencia de un forward, que difiere toda la ganancia o prdida al vencimiento, los futuros se marcan a mercado al precio oficial de liquidacin de cada da hbil. Cada jornada, la bolsa calcula el precio de cierre de ese contrato (settlement price) y ajusta (acredita o debita) las cuentas de margen de los participantes segn la variacin desde el da anterior. Esto garantiza que las prdidas se materialicen y se cobren de inmediato, y las ganancias queden disponibles para entregar o reinvertir.

 **Mrgenes y llamadas de margen (margin call)**
 Para abrir una posicin en futuros se exige depositar un margen inicial (initial margin), calculado para cubrir las prdidas mximas esperables en un perodo de extremo estrs (p. ej., un da de alta volatilidad). Adems, existe un margen de mantenimiento (maintenance margin): si la cuenta de margen cae por debajo de este nivel debido a prdidas en la posicin, la bolsa emite un margin call, obligando al participante a reponer fondos hasta el nivel del margen inicial. Si no lo hace en el plazo sealado, la posicin se puede liquidar automticamente para cubrir el dficit.

 **Liquidez y transparencia de precios**
 La existencia de una bolsa electrnica con numerosos participantes (brokers, hedgers, especuladores, market makers) hace que los contratos ms populares (e.g., futuros de S\&P 500, petrleo ligero, eurodlar) tengan un mercado secundario altamente lquido: spreads bidask reducidos, gran volumen diario en distintos vencimientos y profundos carns de rdenes.

 **Facilidad para coberturas y acceso al apalancamiento**
 Empresas e inversores institucionales usan futuros para cubrir posiciones en activos subyacentes (p. ej., exposicin a tasas de inters, commodities o divisas) sin necesidad de inmovilizar el capital equivalente a comprar o vender fsicamente el activo. Al exigir solo un margen (normalmente un 3  10 % del nocional), los participantes obtienen apalancamiento: con un depsito relativamente pequeo en efectivo pueden controlar una exposicin grande.

#### 3. Riesgo de contraparte y funcin de la **clearing house**

 **Riesgo de contraparte en mercados OTC vs. bolsa**
 En un contrato forwards (OTC), las dos partes acuerdan bilateralmente todas las condiciones. Si una parte entra en insolvencia antes de la fecha de entrega, la contraparte restante asume la prdida correspondiente al valor de mercado del contrato en ese momento. Ese riesgo puede ser significativo si el mercado es voltil y la exposicin nocional es alta. Adems, calcular el Credit Valuation Adjustment (CVA) para valorar el riesgo de crdito de la contraparte aade complejidad cuantitativa, pues requiere estimar la probabilidad de default y la prdida esperada dado el default, usando modelos estructurales (e.g., Merton) o reducidos (p. ej., intensidades de default).

 **Clearing house en contratos de futuros**
 La cmara de compensacin (clearing house) es un ente cuyo objetivo principal es intermediar entre comprador y vendedor para cada transaccin. Cuando un participante abre posicin, la clearing house asume el rol de contraparte para ambas partes. Esto se consigue mediante dos mecanismos fundamentales:



**Margen inicial (Initial Margin)**: depsito en efectivo o valores altamente liquidados (e.g., T-bills, bonos gubernamentales) que sirve para cubrir posibles prdidas si el precio se mueve adversamente el mismo da de la apertura. Este margen se calcula habitualmente con mtodos basados en Value at Risk (VaR) o Extreme Value Theory (EVT), definiendo un nivel de confianza (p. ej., 99 %) y un horizonte de liquidacin (p. ej., 1 da de mercado).

**Margen de mantenimiento (Maintenance Margin)**: si las prdidas acumuladas tras el mark to market diario reducen el saldo de la cuenta por debajo de este nivel, la cmara emite un margin call exigiendo reponer hasta el nivel del margen inicial.

 **Proceso de liquidacin en la clearing house**

1. Al cierre de cada jornada, la bolsa publica el precio de liquidacin del contrato de futuros (settlement price).
2. Se calcula la ganancia o prdida diaria para cada posicin:

$$
\Delta PnL_t = (P_t^{\text{settlement}} - P_{t-1}^{\text{settlement}}) \times \text{tamao del contrato} \times \text{nmero de contratos}
$$


3. Se ajustan las cuentas de margen: si $\Delta PnL_t > 0$, se abona a la cuenta; si $\Delta PnL_t < 0$, se debita.  
4. Si, tras el dbito, la cuenta queda por debajo del margen de mantenimiento, se genera un margin call. El participante debe depositar efectivo o valores elegibles en un plazo determinado (normalmente 24 horas hbiles). Si no lo realiza, la misma clearing house puede liquidar (cerrar) la posicin al precio ms cercano disponible, para cubrir el dficit.  


 **Ventajas del sistema de la clearing house**
 **Eliminacin efectiva del riesgo bilateral**: dado que la cmara garantiza el cumplimiento, no vale la pena intentar estimar CVA/ DVA (Debit Valuation Adjustment) para cada contraparte individual.
 **Estabilidad financiera sistmica**: al existir mrgenes diarios y procedimientos de liquidacin continua, se limita la acumulacin de grandes exposiciones vencidas.
 **Transparencia**: la bolsa informa regularmente de los balances de posiciones abiertas, volmenes de negociacin y niveles de margen requeridos, lo que facilita a supervisores regulatorios (p. ej., CFTC en EE. UU., ESMA en la UE) monitorear riesgos sistmicos potenciales.

#### 4. **Mark to Market**, **Margin Account** y **Margin Call**

**4.1 Mark to Market (Marcacin a Mercado)**



La marcacin a mercado consiste en revalorar diariamente todas las posiciones en futuros al precio oficial de liquidacin del da ($P_t^{\text{settlement}}$). Matemticamente, para una posicin larga de $N$ contratos, cada uno de tamao $L$ (el nocional por contrato), la ganancia (o prdida) en el da $t$ es:

$$
\Delta PnL_t^{\text{largo}}
= \bigl(P_t^{\text{settlement}} - P_{t-1}^{\text{settlement}}\bigr)\times L \times N.
$$

Para una posicin corta, la ganancia diaria se invierte de signo:  


$$
\Delta PnL_t^{\text{corto}}
= \bigl(P_{t-1}^{\text{settlement}} - P_t^{\text{settlement}}\bigr)\times L \times N.
$$

 El mark to market garantiza que las prdidas no se acumulen sin cubrirse. Cada da se ejecuta el siguiente flujo:  
  a) Se determina el precio de liquidacin $P_t^{\text{settlement}}$.  
  b) Se calculan $\Delta PnL_t$ para cada posicin.  
  c) Se actualiza la margin account del participante (vase punto 4.2).  


**4.2 Margin Account (Cuenta de Margen)**

Es la cuenta colateral que cada participante abre en la cmara de compensacin al entrar en una posicin de futuros. Consta de:
 **Margen Inicial (Initial Margin)**: depsito mnimo que se exige al abrir la posicin. Calculado para cubrir el peor escenario de prdidas durante un perodo de liquidacin (look-back period) bajo niveles de confianza predefinidos. Tcnicamente, suele estimarse con un mtodo de Valor en Riesgo (VaR) del 99 % para movimientos extremos de precios en un da de estrs.
 **Margen de Mantenimiento (Maintenance Margin)**: nivel inferior al que la cuenta no puede caer tras la liquidacin diaria. Suele fijarse como un porcentaje (80  90 %) del margen inicial.

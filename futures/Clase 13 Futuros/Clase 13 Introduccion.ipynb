*Rehypothecation was a common practice until 2007, but hedge funds became much more wary about it in the wake of the Lehman Brothers collapse and subsequent credit crunch in 2008-09. #### 1. Principales características de los contratos de futuros

• **Estandarización y negociación en bolsa organizada**
– Cada contrato de futuros especifica de antemano el activo subyacente (p. ej., barriles de petróleo Brent, toneladas métricas de trigo, divisa EUR/USD), el tamaño del lote (e.g., 1 lote = 1 000 bushels de maíz), la calidad o grado mínimo del activo, las fechas de vencimiento (normalmente uno o varios meses fijos al año) y el mecanismo de liquidación (física o cash­settlement). Esta estandarización, impuesta por la bolsa (p. ej., CME, ICE, Euronext), permite que cualquier participante pueda entrar o salir de una posición sin necesidad de negociar condiciones adicionales.

• **Intervención de la cámara de compensación (clearing house)**
– La cámara de compensación actúa como contraparte central para todas las transacciones: cuando alguien abre posición compradora, la cámara se convierte en su contraparte vendedora, y viceversa. Esto elimina el riesgo directo de impago entre dos contrapartes cualquiera, ya que la cámara garantiza el cumplimiento incluso si alguna de las partes originales no pudiera cubrir sus obligaciones.

• **Liquidación diaria (mark to market)**
– A diferencia de un forward, que difiere toda la ganancia o pérdida al vencimiento, los futuros se “marcan a mercado” al precio oficial de liquidación de cada día hábil. Cada jornada, la bolsa calcula el precio de cierre de ese contrato (settlement price) y ajusta (acredita o debita) las cuentas de margen de los participantes según la variación desde el día anterior. Esto garantiza que las pérdidas se materialicen y se cobren de inmediato, y las ganancias queden disponibles para entregar o reinvertir.

• **Márgenes y llamadas de margen (margin call)**
– Para abrir una posición en futuros se exige depositar un “margen inicial” (initial margin), calculado para cubrir las pérdidas máximas esperables en un período de extremo estrés (p. ej., un día de alta volatilidad). Además, existe un “margen de mantenimiento” (maintenance margin): si la cuenta de margen cae por debajo de este nivel debido a pérdidas en la posición, la bolsa emite un “margin call”, obligando al participante a reponer fondos hasta el nivel del margen inicial. Si no lo hace en el plazo señalado, la posición se puede liquidar automáticamente para cubrir el déficit.

• **Liquidez y transparencia de precios**
– La existencia de una bolsa electrónica con numerosos participantes (brokers, hedgers, especuladores, market makers) hace que los contratos más populares (e.g., futuros de S\&P 500, petróleo ligero, eurodólar) tengan un mercado secundario altamente líquido: spreads bid‐ask reducidos, gran volumen diario en distintos vencimientos y profundos carnés de órdenes.

• **Facilidad para coberturas y acceso al apalancamiento**
– Empresas e inversores institucionales usan futuros para cubrir posiciones en activos subyacentes (p. ej., exposición a tasas de interés, commodities o divisas) sin necesidad de inmovilizar el capital equivalente a comprar o vender físicamente el activo. Al exigir solo un margen (normalmente un 3 – 10 % del nocional), los participantes obtienen apalancamiento: con un depósito relativamente pequeño en efectivo pueden controlar una exposición grande.

#### 3. Riesgo de contraparte y función de la **clearing house**

• **Riesgo de contraparte en mercados OTC vs. bolsa**
– En un contrato forwards (OTC), las dos partes acuerdan bilateralmente todas las condiciones. Si una parte entra en insolvencia antes de la fecha de entrega, la contraparte restante asume la pérdida correspondiente al valor de mercado del contrato en ese momento. Ese riesgo puede ser significativo si el mercado es volátil y la exposición nocional es alta. Además, calcular el “Credit Valuation Adjustment” (CVA) para valorar el riesgo de crédito de la contraparte añade complejidad cuantitativa, pues requiere estimar la probabilidad de default y la pérdida esperada dado el default, usando modelos estructurales (e.g., Merton) o reducidos (p. ej., intensidades de default).

• **Clearing house en contratos de futuros**
– La cámara de compensación (clearing house) es un ente cuyo objetivo principal es intermediar entre comprador y vendedor para cada transacción. Cuando un participante abre posición, la clearing house asume el rol de contraparte para ambas partes. Esto se consigue mediante dos mecanismos fundamentales:



**Margen inicial (Initial Margin)**: depósito en efectivo o valores altamente liquidados (e.g., T-bills, bonos gubernamentales) que sirve para cubrir posibles pérdidas si el precio se mueve adversamente el mismo día de la apertura. Este margen se calcula habitualmente con métodos basados en Value at Risk (VaR) o Extreme Value Theory (EVT), definiendo un nivel de confianza (p. ej., 99 %) y un horizonte de liquidación (p. ej., 1 día de mercado).

**Margen de mantenimiento (Maintenance Margin)**: si las pérdidas acumuladas tras el mark to market diario reducen el saldo de la cuenta por debajo de este nivel, la cámara emite un “margin call” exigiendo reponer hasta el nivel del margen inicial.

• **Proceso de liquidación en la clearing house**

1. Al cierre de cada jornada, la bolsa publica el “precio de liquidación” del contrato de futuros (settlement price).
2. Se calcula la ganancia o pérdida diaria para cada posición:

$$
\Delta PnL_t = (P_t^{\text{settlement}} - P_{t-1}^{\text{settlement}}) \times \text{tamaño del contrato} \times \text{número de contratos}
$$


3. Se ajustan las cuentas de margen: si $\Delta PnL_t > 0$, se abona a la cuenta; si $\Delta PnL_t < 0$, se debita.  
4. Si, tras el débito, la cuenta queda por debajo del margen de mantenimiento, se genera un margin call. El participante debe depositar efectivo o valores elegibles en un plazo determinado (normalmente 24 horas hábiles). Si no lo realiza, la misma clearing house puede liquidar (cerrar) la posición al precio más cercano disponible, para cubrir el déficit.  


• **Ventajas del sistema de la clearing house**
– **Eliminación efectiva del riesgo bilateral**: dado que la cámara garantiza el cumplimiento, no vale la pena intentar estimar CVA/ DVA (Debit Valuation Adjustment) para cada contraparte individual.
– **Estabilidad financiera sistémica**: al existir márgenes diarios y procedimientos de liquidación continua, se limita la acumulación de grandes exposiciones vencidas.
– **Transparencia**: la bolsa informa regularmente de los balances de posiciones abiertas, volúmenes de negociación y niveles de margen requeridos, lo que facilita a supervisores regulatorios (p. ej., CFTC en EE. UU., ESMA en la UE) monitorear riesgos sistémicos potenciales.

#### 4. **Mark to Market**, **Margin Account** y **Margin Call**

**4.1 Mark to Market (Marcación a Mercado)**



La marcación a mercado consiste en revalorar diariamente todas las posiciones en futuros al precio oficial de liquidación del día ($P_t^{\text{settlement}}$). Matemáticamente, para una posición larga de $N$ contratos, cada uno de tamaño $L$ (el nocional por contrato), la ganancia (o pérdida) en el día $t$ es:

$$
\Delta PnL_t^{\text{largo}}
= \bigl(P_t^{\text{settlement}} - P_{t-1}^{\text{settlement}}\bigr)\times L \times N.
$$

Para una posición corta, la ganancia diaria se invierte de signo:  


$$
\Delta PnL_t^{\text{corto}}
= \bigl(P_{t-1}^{\text{settlement}} - P_t^{\text{settlement}}\bigr)\times L \times N.
$$

– El mark to market garantiza que las pérdidas no se acumulen sin cubrirse. Cada día se ejecuta el siguiente flujo:  
  a) Se determina el precio de liquidación $P_t^{\text{settlement}}$.  
  b) Se calculan $\Delta PnL_t$ para cada posición.  
  c) Se actualiza la “margin account” del participante (véase punto 4.2).  


**4.2 Margin Account (Cuenta de Margen)**

Es la cuenta colateral que cada participante abre en la cámara de compensación al entrar en una posición de futuros. Consta de:
• **Margen Inicial (Initial Margin)**: depósito mínimo que se exige al abrir la posición. Calculado para cubrir el peor escenario de pérdidas durante un período de liquidación (look-back period) bajo niveles de confianza predefinidos. Técnicamente, suele estimarse con un método de Valor en Riesgo (VaR) del 99 % para movimientos extremos de precios en un día de estrés.
• **Margen de Mantenimiento (Maintenance Margin)**: nivel inferior al que la cuenta no puede caer tras la liquidación diaria. Suele fijarse como un porcentaje (80 – 90 %) del margen inicial.

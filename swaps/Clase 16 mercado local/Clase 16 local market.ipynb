### 2. Fundamentos del SPC

**1. Swap de Tasas de Interés**:

Un **swap de tasas de interés** es un contrato OTC bilateral en el que la contraparte “A” paga una tasa fija sobre un capital nocional y la contraparte “B” paga una tasa variable (o flotante) sobre ese mismo capital y plazo. En la práctica, los pagos de intereses se “netean”, de modo que solo la diferencia se liquida en cada fecha de pago. Por ejemplo, en un swap a 5 años con nocional:
- US\$ 5 000 000
- Si la tasa fija es 9,5 % 
- Y la LIBOR al primer vencimiento es 8 %

- “A” pagará US$ 237 500 y “B” US$ 200 000, por lo que “A” netea un pago de US$ 37 500 a “B” .


### 3. Contexto del mercado chileno

En Chile los swaps de tasas en moneda local se denominan **Swap Promedio Cámara en pesos (SPC CLP)** y **Swap Promedio Cámara en UF (SPC UF)**. En un SPC CLP, “A” paga una tasa fija en pesos y “B” paga la tasa interbancaria diaria promedio **(PTID)** sobre el mismo nocional y plazo; en SPC UF ambas tasas son expresadas en UF . Los pagos de intereses suelen liquidarse semestralmente, y el capital se ajusta al cierre del contrato siguiendo una estructura similar a la de los bonos del Banco Central.




* Modalidades: **SPC CLP** y **SPC UF**, con plazos típicos a 6 m, 1, 2, 3, 5 años (UF hasta 20 años) .
* Principales agentes: \~10 bancos locales, hedge funds y bancos extranjeros (Morgan Stanley, Goldman Sachs, etc.) que utilizan SPC como vía de entrada al mercado local libre de riesgo .
* Volúmenes: hacia agosto 2005, \~\$375 mil millones mensuales.

#### Índice de Cámara Promedio (ICP)


La tasa variable del SPC se basa en el **Promedio de la Tasa Interbancaria Diaria (PTID)**, publicado por el Banco Central. Para convertir la PTID anual a una tasa mensual (TCIPm), hasta octubre 2006 se usaba compounding:

$$
\text{TCIPm} = \bigl[(1 + \tfrac{\text{PTID}}{100})^{30/360} - 1\bigr]\times100,
$$


En octubre de 2002 se publicó en el Diario Oficial el
“Reglamento Indice Cámara Promedio”, aprobado por la Asociación de Bancos, el cual fue definido 
como “un indicador que busca representar el costo de fondos equivalente de financiar una posición a 
la tasa overnight, utilizando para esto la Tasa Cámara Interbancaria
Promedio informada por el Banco Central de Chile.”.


y desde noviembre 2006 en forma simple:

$$
\text{TCIPm} = \tfrac{\text{PTID}}{12}.
$$


Con la TCIPm diaria se construye el **Índice Cámara Promedio (ICP)** recurriendo a

$$
\text{ICPi} = \text{ICPi-1}\times\bigl(1 + \tfrac{\text{TCIPm}_{i-1}\times N}{3000}\bigr),
$$



### 4. Aplicaciones principales y Casos

* **Cobertura de tasa de interés:** Un banco con pasivo fijo a dos años puede entrar a un SPC CLP fijo/flotante para convertirlo en pasivo variable, neutralizando riesgo ante disminuciones de tasas .
* **Especulación:** Si la bid del SPC CLP a 6 m es 5,4 % y se anticipa una baja a 5 %, un agente entra pagando flotante y recibiendo fijo, obteniendo la diferencia al vencimiento .
* **Arbitraje y swap spread:** El **swap spread** es la diferencia entre la tasa SPC y la del bono equivalente del Banco Central (BCP). Este crece con el plazo y refleja riesgo crédito/liquidez. Por ejemplo, al 22 junio 2006, el spread a 2 años fue 16 pb y a 5 años 46 pb .



Durante la primera mitad de 2005 el spread a 4 años osciló entre 20 y 40 pb; desde mediados de 2005 y durante 2006 subió y se mantuvo entre 40 y 70 pb, afectado por expectativas de alza de tasas . Se observa que el spread en UF a 4 años (52 pb) superó al de pesos (30 pb) en junio 2006, debido al riesgo inflacionario incorporado en UF .


#### Casos

* **Caso 1: Concesionaria emite bono en exterior**

  1. La concesionaria emite un bono en US$ y celebra un CCS UF fijo US$ fijo con banco A.
  2. Banco A, con activo UF fijo y pasivo US$ fijo, cubre riesgo con:

     * SPC UF fijo / UF flotante con banco B
     * CCS UF flotante / US$ flotante con banco C
     * IRS US$ flotante / US$ fijo con banco D
  3. Así, banco A neutraliza exposición de tasas y moneda en tres pasos .

La concesionaria efectúa una operación de Cross Currency Swap (CCS) UF fijo/US$ fijo (paga tasa fija en UF y recibe tasa fija en dólares) con el banco A, con  lo cual cubre su pasivo en dólares y genera un pasivo en UF. Por su parte, el banco A, que queda  con un activo a tasa fija en UF y un pasivo a tasa fija en dólares, desea cubrir esta operación 
(cubrir su riesgo de tasa y moneda) para lo cual efectúa tres operaciones adicionales con otros bancos (o con un mismo banco). 

En primer lugar:
- El banco A realiza un SPC UF fijo/UF flotante con un banco B; 
- Luego, el banco A, efectúa un CCS UF flotante/US$ flotante con un banco C;
- Finalmente, el banco A, realiza un Swap de Tasas (IRS) US$ flotante/US$ fijo con un banco D. 

Con estas tres últimas operaciones, el banco A cubre el riesgo financiero, de tasas y moneda, que le 
generó la operación con la concesionaria.


* **Caso 2: Banco extranjero invierte en SPC CLP**
  Con tasas bid/ask a 5 años de 6,75/6,85 %, un banco global que pronostica alza entra captando a fijo (6,85 %) y colocando flotante; la contraparte local hace la operación inversa aprovechando su expectativa contraria .



**5. Ejemplos numéricos detallados**

* **Ejemplo SPC CLP (183 días, nocional \$2 000 000 000)**

  * Tasa fija: 2,8 % anual → 0,23 % mensual
  * PTID promedio: 3,0 % anual → 0,25 % mensual
  * Intereses A → B: \$2 000 000 000×0,23 %×(183/30)=\$28 060 000
  * Intereses B → A: \$2 000 000 000×0,25 %×(183/30)=\$30 500 000
  * Neteo: B paga a A \$2 440 000 .
* **Ejemplo SPC UF (88 días, nocional UF 616 757,42)**

  * Tasa fija UF: 0,40 % anual
  * PTID UF efectivo: 0,47 % anual
  * Intereses A → B: UF 603,05
  * Intereses B → A: UF 708,59
  * Neteo: B paga a A UF 105,54 (≈\$1 726 478) .
